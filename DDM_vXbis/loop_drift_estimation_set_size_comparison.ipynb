# Parameters

```python
SET_SIZES_LIST = (10,20,50,100)
MAIN_FOLDER_PATH = '/home/david/Documents/code/loop_drift_estimation_output/'
```

# Loading estimations

```python

estimated_drift_per_states_number_list = []

for set_size in SET_SIZES_LIST:

    with open(f'{MAIN_FOLDER_PATH}/n_{set_size}/estimated_drift_list.pkl', 'rb') as file:
            stacked_recovered_drift_per_states_number = pickle.load(file)

    estimated_drift_per_states_number_list.append(stacked_recovered_drift_per_states_number)
```

```python
example_drift = 2
example_state = 1

fig=plt.figure(figsize=(4, 2.5), dpi=300, constrained_layout=False, facecolor='w')
gs = fig.add_gridspec(1, 1)
row = gs[:].subgridspec(1, 1, hspace=0.5)

ax = plt.subplot(row[0,0])

mean_list = []
std_list = []

for i, set_size in enumerate(SET_SIZES_LIST):

    estimated_drift_list = estimated_drift_per_states_number_list[i][example_drift][example_state]

    mean_value = np.mean(estimated_drift_list)
    std_value = np.std(estimated_drift_list)

    mean_list.append(mean_value)
    std_list.append(std_value)

    ax.scatter(np.ones(len(estimated_drift_list))*set_size, estimated_drift_list, alpha=0.2, marker='+')

ax.errorbar(SET_SIZES_LIST, mean_list, std_list, color='k', capsize=5, alpha=0.7)

ax.axhline(0.0525, color='k', linestyle='--')

ax.set_ylim([0,0.15])
ax.set_xlabel('Size of set (in nbr of simulations)')
ax.set_ylabel('Recovered drift')
ax.set_xticks(SET_SIZES_LIST)
ax.set_title(f'{example_state+2} states', fontsize=7)
ax.legend()
```
